## '?' olan verileri null yaptım

```python
dataFrameTrain = dataFrameTrain.replace('?',np.nan)
dataFrameTest = dataFrameTest.replace('?',np.nan)
```

## Yaş ve cinsiyeti null olan verileri çıkarttım

```python
dataFrameTrain.dropna(subset=["age"], inplace=True)
dataFrameTrain.dropna(subset=["sex"], inplace=True)
dataFrameTest.dropna(subset=["sex"], inplace=True)
```

## Null olan verileri kolon ortalamasına göre doldurdum

```python
meanTSH = (dataFrameTrain["TSH"]+dataFrameTest['TSH']).mean()
meanT3 = (dataFrameTrain["T3"]+dataFrameTest['T3']).mean()
meanTT4 = (dataFrameTrain["TT4"]+dataFrameTest['TT4']).mean()
meanT4U = (dataFrameTrain["T4U"]+dataFrameTest['T4U']).mean()
meanFTI = (dataFrameTrain["FTI"]+dataFrameTest['FTI']).mean()
```

```python
values = {'TSH':meanTSH, 'T3':meanT3, 'TT4':meanTT4, 'T4U':meanT4U, 'FTI':meanFTI}
dataFrameTrain=dataFrameTrain.fillna(value=values)
dataFrameTest = dataFrameTest.fillna(value=values)
```

## Cinsiyet ve true, false ola ndeğerleri 0-1 lere çevirdim

```python
dataFrameTrain["sex"] = dataFrameTrain["sex"].replace('F',0)
dataFrameTrain["sex"] = dataFrameTrain["sex"].replace('M',1)
dataFrameTest["sex"] = dataFrameTest["sex"].replace('F',0)
dataFrameTest["sex"] = dataFrameTest["sex"].replace('M',1)
```

```python
dataFrameTrain = dataFrameTrain.replace('f',0)
dataFrameTrain = dataFrameTrain.replace('t',1)
dataFrameTest = dataFrameTest.replace('f',0)
dataFrameTest = dataFrameTest.replace('t',1)
```

## Train ve Test datalarını x,y olarak ayırıyorum. Ayrıca "referral source" kolonunu çıkarttım

```python
y_train = dataFrameTrain["class"].values
x_train = dataFrameTrain.drop(["referral source","class"],axis = 1).values

y_test = dataFrameTest["class"].values
x_test = dataFrameTest.drop(["referral source","class"],axis = 1).values
```

## Class kolonunu dummy codding ile kategorik hale getiriyorum

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(y_train)
encoded_Y = encoder.transform(y_train)

encoder_test =LabelEncoder()
encoder_test.fit(y_test)
encoded_Y_test = encoder.transform(y_test)
```

```python
import tensorflow as tf
from keras.utils import np_utils
dummy_y = np_utils.to_categorical(encoded_Y)
dummy_y_test =np_utils.to_categorical(encoded_Y_test)
```

## Min-Max normalizasyonu yapıyorum

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minMax = MinMaxScaler()
```

```python
minMax.fit(x_train)
```

```python
x_train = minMax.transform(x_train)
```

```python
x_test = minMax.transform(x_test)
```

# MLP modelini oluşturuyorum

```python
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
optimizer = keras.optimizers.Adam(learning_rate=0.002)
```

### loss ve val_loss değerlerinin beraber azalmasını isteidğimiz için dropout kullanmayı denedim

```python
model = Sequential()

model.add(Dense(14,input_dim=27,activation = "tanh"))
model.add(Dropout(0.3))

model.add(Dense(units=14,activation = "tanh"))
model.add(Dropout(0.3))

model.add(Dense(units=3,activation = "softmax"))

model.compile(loss="categorical_crossentropy",optimizer = optimizer,metrics=['accuracy'])
```

## Modeli fit ediyorum

### Early stopping kullanıyorum epoch değerini dinamik ayarlanmasını ve vall_loss un en düşük olduğu yerde sonlanmasını istiyorum.

```python
# model.fit(x=x_train, y=y_train, epochs = 50, validation_data = (x_test,y_test),batch_size=30)

earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
model.fit(x=x_train, y=dummy_y, epochs = 300, validation_data = (x_test,dummy_y_test), verbose = 1, callbacks=[earlyStopping],batch_size=30)
```

## Grafik

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
import matplotlib.pyplot as plt
modelKaybi.plot()
```

```python
test_predicted = model.predict(x_test)
test_predicted
```

```python
train_predicted = model.predict(x_train)
train_predicted
```

# İstenen Tablo

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
test_predicted = np.argmax(test_predicted, axis=1)
test_y = np.argmax(dummy_y_test, axis=1)
classNames=["negative","primary_hypothyroid","compensated_hypothyroid"]
test_table = classification_report(test_y,test_predicted,target_names=classNames,labels=[0,1,2],digits=3,zero_division=0,output_dict=True)
print(classification_report(test_y,test_predicted,target_names=classNames,labels=[0,1,2],digits=3,zero_division=0))
```

```python
train_predicted = np.argmax(train_predicted, axis=1)
train_y = np.argmax(dummy_y, axis=1)
classNames=["negative","primary_hypothyroid","compensated_hypothyroid"]
train_table = classification_report(train_y,train_predicted,target_names=classNames,labels=[0,1,2],digits=3,zero_division=0,output_dict=True)
print(classification_report(train_y,train_predicted,target_names=classNames,labels=[0,1,2],digits=3,zero_division=0))
```
